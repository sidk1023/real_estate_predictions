```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import json
```

```python
def makaan_scraper_v2():
    '''scrapes the popular real estate website makaan.com to get data such as coordinates, property type, locality, area'''
    import requests
    from bs4 import BeautifulSoup
    import numpy as np
    import pandas as pd
    url = 'https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city'
    response = requests.get(url)
    if response.ok:
        soup = BeautifulSoup(response.text, "lxml")
        pagenav = soup.select(".pagination > :nth-of-type(1) >:nth-of-type(7)>:nth-of-type(1)")
        num_pages = int(pagenav[0].text)
    lats=[]
    longs=[]
    seller_type=[]
    no_beds=[]
    layout_type=[]
    property_type=[]
    locality_type=[]
    area_value=[]
    area_unit=[]
    furnishing=[]
    prices = []
    #n_bathrooms=[]
    for i in range(1,num_pages):
        url = 'https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city' + '?page='+str(i)
        response = requests.get(url)
        if response.ok:
            soup = BeautifulSoup(response.text, "lxml")

            latitudes = [soup.find_all(attrs={"itemprop": "latitude"})] 
            lats.extend([str(latitude)[15:25] for latitude in latitudes[0]]) 

            longitudes = [soup.find_all(attrs={"itemprop": "longitude"})] 
            longs.extend([str(longitude)[15:25] for longitude in longitudes[0]]) 

            sellers = soup.find_all(attrs={"class": "seller-type"})
            seller_type.extend([seller.text for seller in sellers])

            titles= soup.select('a.typelink')
            titles = [title.text.split() for title in titles]

            no_beds.extend([title[0] for title in titles])
            layout_type.extend([title[1] for title in titles])
            property_type.extend([' '.join(title[2:]) for title in titles])

            localities = soup.find_all(attrs={"itemprop": "addressLocality"})
            locality_type.extend([locality.text for locality in localities])

            areas = soup.find_all(attrs={"class": "size"})
            area_value.extend([area.text for area in areas])

            units = [soup.select(".size ~ .lbl")] 
            area_unit.extend([unit.text for unit in units[0]])

            ftypes = soup.find_all(attrs={"class": "hcol w44"})
            furnishing.extend([ftype.text for ftype in ftypes])
            
            pricex = soup.find_all(attrs={"itemprop": "offers"})
            prices.extend([price.text for price in pricex])
            #bathrooms = soup.find_all(attrs={"title": "bathrooms"})
            #n_bathrooms.extend([broom.text for broom in bathrooms])
            print(i)    
    data=[lats,longs,seller_type,no_beds,layout_type,property_type,locality_type,area_value,area_unit,furnishing,prices]
    columns = ('latitude','longitude','seller','bedrooms','layout','property_type','locality','area','units','furnishing','rent')
    df = pd.DataFrame(data,columns).transpose() 
    return df
```

```python
def makaan_scraper_v1():
    '''scrapes the popular real estate website makaan.com to get data such as coordinates, property type, locality, area'''
    import requests
    from bs4 import BeautifulSoup
    import numpy as np
    import pandas as pd
    import json
    url = 'https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city'
    response = requests.get(url)
    if response.ok:
        soup = BeautifulSoup(response.text, "lxml")
        pagenav = soup.select(".pagination > :nth-of-type(1) >:nth-of-type(7)>:nth-of-type(1)")
        num_pages = int(pagenav[0].text)
        mainlist=[]
    for i in range(1,num_pages):
        url1 = 'https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city' + '?page='+str(i)
        response = requests.get(url1)
        if response.ok:
            soup = BeautifulSoup(response.text, "lxml")
            cards = soup.select(".cardWrapper > script")
            for j in range(len(cards)):
                attrs = json.loads(cards[j].text)
                mylist=[attrs['latitude'],attrs['longitude'],attrs['bedrooms'],attrs['sellerType'],attrs['propertyType'],attrs['localityName'],attrs['suburbName'],attrs['size'].split()[0],('').join(attrs['size'].split()[1:]),attrs['price']]
                mainlist.append(mylist)
        print(i)        

            
    
    columns = ('latitude','longitude','bedrooms','seller','property_type','locality','suburb','area','units','rent')
    df = pd.DataFrame(mainlist,columns=columns)
    return df
```

```python
df = makaan_scraper_v1()
df.to_csv('hyd_rents')
```
